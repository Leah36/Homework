```python
#Calculate the total number of schools
schools = schools_df["school_name"].count()
schools

#Calculate the total number of students
students = students_df["student_name"].count()
students

#Calculate the total budget
total_budget = schools_df["budget"].sum()
total_budget

#Calculate the average math score
math_score = students_df["math_score"].mean()
math_score

#Calculate the average reading score
reading_score = students_df["reading_score"].mean()
reading_score

#Calculate the overall passing rate(overall average score),ie (avg math score + avg reading score)/2
overall_average_score = (math_score + reading_score)/2
overall_average_score

#Calculate the percentage of students with a passing math score (70 or greater)
students_df["passing_math"] = students_df["math_score"] >= 70
students_df["passing_reading"] = students_df ["reading_score"] >=70
percent_passing_math = ((students_df["passing_math"]).mean()) *100

#Calculate the percentage of students with a passing reading score (70 or greater)
percent_passing_reading = ((students_df["passing_reading"]).mean())*100

#Calculate overall percentage
overall_passing_rate = (percent_passing_math + percent_passing_reading)/2
overall_passing_rate

district_results=[{"schools": schools, "students":students,
                  "total_budget":total_budget, "math_score": round (math_score,2),
                  "%Passing Math":round(percent_passing_math,2),
                  "%Passing Reading": round(percent_passing_reading,2),
                  "Overall Passing Rate": round(overall_passing_rate,2)}]
district_summary_table= pd.DataFrame(district_results)

district_summary= [[ "schools"," students", "total budget", "math score", "reading score", "%Passing Math", "%Passing Reading", "%_Overall Passing Rate"]]

# #formatting cells
district_summary_format=({"budget": "${:,.2f}", 
                        " reading score": "{:.1f}", 
                        " math score": "{:.1f}", 
                        "% Passing Math": "{:.1%}", 
                        "% Passing Reading": "{:.1%}", 
                        "Overall Passing Rate": "{:.1%}"})

# print out district summary format
district_results
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Reorganize schools data table alphabetically, rename columns, and set index

schools_sort = schools_df.sort_values("school_name", inplace=False)
schools_rename_df = schools_sort.rename(columns={"school_name": "", "type": "School Type", "size": "Total Students", "budget": "Total School Budget"}) 
schools_rename_df = schools_rename_df.set_index("")

#Per student budget
per_student_budget = schools_rename_df["Total School Budget"]/schools_rename_df["Total Students"]
schools_rename_df["Per Student Budget"] = per_student_budget

# Average math scores
school_group = students_df.groupby(["school_name"])
math_average = school_group["math_score"].mean()
schools_rename_df["math score"] = math_score

# Average reading scores   
reading_average = school_group["reading_score"].mean()
schools_rename_df["reading score"] = reading_score

# % passing math
math_pass = students_df.loc[students_df["math_score"] >= 70]
math_pass_group = math_pass.groupby(["school_name"])
math_pass_count = math_pass_group["math_score"].count()
perc_math_pass = math_pass_count/schools_rename_df["Total Students"] * 100
schools_rename_df["% Passing Math"] = perc_math_pass

# % passing reading
reading_pass = students_df.loc[students_df["reading_score"] >= 70]
reading_pass_group = reading_pass.groupby(["school_name"])
reading_pass_count = reading_pass_group["reading_score"].count()
perc_reading_pass = reading_pass_count/schools_rename_df["Total Students"] * 100
schools_rename_df["% Passing Reading"] = perc_reading_pass

 # % overall passing math
overall_pass_rate = (schools_rename_df["% Passing Math"] + schools_rename_df["% Passing Reading"])/2
schools_rename_df["% Overall Passing Rate"] = overall_pass_rate

# Format columns
schools_rename_df["Total School Budget"] = schools_rename_df["Total School Budget"].map("${:,.2f}".format)
schools_rename_df["Per Student Budget"] = schools_rename_df["Per Student Budget"].map("${:,.2f}".format)

# Print out school summary table
schools_rename_df
```

## Math Scores by Grade

```python
 # Average math score at each grade level for each school

# ninth_grade = school_complete[school_complete["grade"] == "9th"].groupby ("school_name").mean()
# ["math_score"]

ninth_grade = students_df.loc[students_df["grade"] == "9th"]
group_ninth = ninth_grade.groupby(["school_name"])
ninth_avg_math = group_ninth["math_score"].mean()

# Tenth Grade
tenth_grade = students_df[students_df["grade"] == "10th"]
group_tenth = tenth_grade.groupby(["school_name"])
tenth_avg_math = group_tenth["math_score"].mean()

# Eleventh Grade
eleventh_grade = students_df[students_df["grade"] == "11th"]
group_eleventh = eleventh_grade.groupby(["school_name"])
eleventh_avg_math = group_eleventh["math_score"].mean()

# Twelfth Grade
twelfth_grade = students_df[students_df["grade"] == "12th"]
group_twelfth = twelfth_grade.groupby(["school_name"])
twelfth_avg_math = group_twelfth["math_score"].mean()

# Create dataframe
summary_math = pd.DataFrame({"9th": ninth_avg_math})

# Add in remaining grade columns
summary_math["10th"] = tenth_avg_math
summary_math["11th"] = eleventh_avg_math
summary_math["12th"] = twelfth_avg_math

 # Reset index name
summary_math.index.name = ""

 # Print average math summary table by grades
summary_math
```
